### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
Individual_Players = purchase_data['SN'].value_counts()
Total_players = Individual_Players.count()
Total_players_df = pd.DataFrame([{"Total Players":Total_players}])
Total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Number_items = purchase_data['Item ID'].value_counts()
Unique_items = Number_items.count()
Average_price = round(purchase_data['Price'].mean(),2)
Number_of_purchases = Number_items.sum()
Total_Revenue = purchase_data['Price'].sum()

Basic_calculations = pd.DataFrame ([{"Number of Unique Items":Unique_items, "Average Price":Average_price, "Number of Purchases":Number_of_purchases, "Total Revenue":Total_Revenue}])
Basic_calculations["Average Price"]=Basic_calculations["Average Price"].map('${}'.format)
Basic_calculations["Total Revenue"]=Basic_calculations["Total Revenue"].map('${:,}'.format)
Basic_calculations_ordered = Basic_calculations.loc[:,["Number of Unique Items","Average Price","Number of Purchases","Total Revenue"]]
Basic_calculations_ordered
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Maletot = purchase_data.loc[purchase_data["Gender"] == "Male"]
Male = Maletot['SN'].count()
Femaletot = purchase_data.loc[purchase_data["Gender"] == "Female"]
Female = Femaletot['SN'].count()
Othertot = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed"]
Other = Othertot['SN'].count()

PercentageMale = round(Male / Total_players * 100,2)
PercentageFemale = round(Female / Total_players * 100,2)
PercentageOther = round(Other / Total_players * 100,2)

Demo = pd.DataFrame({"":["Male","Female","Other / Non-Disclosed"],"Percentage of Players":[PercentageMale,PercentageFemale,PercentageOther], "Total Count":[Male,Female,Other]})
Demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
AveragepriceMale = round(Maletot['Price'].mean(),2)
AveragepriceFemale = round(Femaletot['Price'].mean(),2)
AveragepriceOther = round(Othertot['Price'].mean(),2)

TotalMale = round(Maletot['Price'].sum(),2)
TotalFemale = round(Femaletot['Price'].sum(),2)
TotalOther = round(Othertot['Price'].sum(),2)

PerpersonMale = round(TotalMale/Male,2)
PerpersonFemale = round(TotalFemale/Female,2)
PerpersonOther = round(TotalOther/Other,2)

Gender = pd.DataFrame({"Gender":["Female","Male","Other / Non-Disclosed"], "Purchase Count":[Female,Male,Other], "Average Purchase Price":[AveragepriceFemale,AveragepriceMale,AveragepriceOther],"Total Purchase Value":[TotalFemale,TotalMale,TotalOther],"Avg Purchase Total per Person":[PerpersonFemale,PerpersonMale,PerpersonOther]})
Gender['Average Purchase Price']=Gender['Average Purchase Price'].map('${:.2f}'.format)
Gender['Avg Purchase Total per Person']=Gender['Avg Purchase Total per Person'].map('${:.2f}'.format)
Gender['Total Purchase Value']=Gender['Total Purchase Value'].map("${:,}".format)
Gender_ordered = Gender.loc[:,["Gender","Purchase Count","Average Purchase Price","Total Purchase Value","Avg Purchase Total per Person"]]
Gender_ordered
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data['Age bin'] = pd.cut(purchase_data['Age'],age_bins,labels = group_names)

groupedbyage = purchase_data.groupby('Age bin')
Totalgroupedbyage = groupedbyage['SN'].count()
Percentagegroupedbyage = round(Totalgroupedbyage/Total_players * 100,2)

Age = pd.DataFrame({"Percentage of Players":Percentagegroupedbyage,"Total Count":Totalgroupedbyage})
Age
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins2 = [9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names2 = ["10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data['Age bin'] = pd.cut(purchase_data['Age'],age_bins2,labels = group_names2)

groupedbyage = purchase_data.groupby('Age bin')

Purchasecount = round(groupedbyage['SN'].count(),2)
Totalpurchaseprice = round(groupedbyage['Price'].sum(),2)
Averagepurchaseprice = round(groupedbyage['Price'].mean(),2)
Averagepurchasepriceperperson = round(Totalpurchaseprice/Purchasecount,2)

purchasingage = pd.DataFrame({"Purchase Count":Purchasecount,"Average Purchase Price":Averagepurchaseprice, "Total Purchase Value":Totalpurchaseprice,"Average Purchase Total per Person":Averagepurchasepriceperperson})
purchasingage['Average Purchase Price']=purchasingage['Average Purchase Price'].map('${}'.format)
purchasingage['Average Purchase Total per Person']=purchasingage['Average Purchase Total per Person'].map('${}'.format)
purchasingage['Total Purchase Value']=purchasingage['Total Purchase Value'].map('${}'.format)

purchasingage_ordered = purchasingage.loc[:,["Purchase Count","Average Purchase Price","Total Purchase Value","Average Purchase Total per Person"]]
purchasingage_ordered
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Items = purchase_data.loc[:,["Item ID","Item Name","Price"]]

number_items= Items.groupby(['Item ID','Item Name'])["Price"].count().rename("number_items").reset_index()
Items_1 = Items.merge(number_items)
Items_1["Total_purchase_value"] = Items_1["Price"] * Items_1["number_items"]


summaryItem = pd.DataFrame({"Item ID":Items_1['Item ID'], "Item Name":Items_1['Item Name'],"Purchase Count":Items_1["number_items"],"Item Price":Items_1['Price'],"Total Purchase Value":Items_1["Total_purchase_value"]})
Items_2 = summaryItem.groupby(['Item ID','Item Name']).first()
sortedItems_2 = Items_2.sort_values(by='Purchase Count',ascending=False)
sortedItems_2['Item Price']=sortedItems_2['Item Price'].map('${:.2f}'.format)
sortedItems_2['Total Purchase Value']=sortedItems_2['Total Purchase Value'].map('${:.2f}'.format)
sortedItems_2_ordered = sortedItems_2.loc[:,["Purchase Count","Item Price","Total Purchase Value"]]
sortedItems_2_ordered.head()
```
